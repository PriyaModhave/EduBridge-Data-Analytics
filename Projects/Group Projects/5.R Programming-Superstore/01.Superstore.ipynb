### Region Wise Total Sales

```r
Regionwise_Profit=df %>% group_by(Region) %>% summarise(total_Profit=sum(Profit))
Regionwise_Profit
```

## Visualization:

### Price per product in different sub_categories

```r
price_prod = ggplot(df,aes(x= Sub.Category,y= sum(Sales)/sum(Quantity),fill = Sub.Category))+ ggtitle("Profit per product") +theme(axis.text.x = element_text(angle=60,size =5 ))+geom_col()
price_prod
```

## Profitable Office in california State

```r
ca = df %>% group_by(Sub.Category) %>% filter(Category== 'Office Supplies' & State == 'California') %>% summarise(Total_Profit = sum(Profit))
ca
ca_plot = ggplot(ca, aes(x= Total_Profit, y= Sub.Category,fill = Sub.Category))+ ggtitle("Proftable Office in california State")+geom_col()
ca_plot
```

## State Wise Sales Analysis

```r
sales_A=ggplot(df,aes(x=State,y=Sales,fill=State))+geom_col()+coord_flip()+theme(legend.position='None')+ggtitle('State wise analysis ')
sales_A
```

### Region Wise Profit Analysis

```r
Region_A=ggplot(df,aes(x=Region,y=Profit,fill=Region))+geom_col()+coord_flip()+theme(legend.position='None')+ggtitle('Region wise analysis ')
Region_A
```
